```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_excel('Online Retail.xlsx')
```

```python
data.head()
```

```python
# Dropping all records with empty values
data.dropna(inplace = True)
data.isna().any()
```

```python
# Removing negative values
data = data[data.Quantity >= 0]
data.describe()
```

```python
user_item_matrix = data.pivot_table(index = 'CustomerID', columns='Description', values='Quantity')
user_item_matrix
```

```python
data.groupby(['Description'])['Quantity'].sum().sort_values(ascending = False).head(15)
```

```python
qty_list = list(data.groupby(['Description'])['Quantity'].sum().sort_values(ascending = False).head(15).index)
qty_list
```

```python
# Let us recommend for a user who bought the tenth most popular item
item = qty_list[14]
item
```

```python
# To get customers who bought item
item_buyers = user_item_matrix[item]
item_buyers.head()
```

```python
items_like_item = user_item_matrix.corrwith(item_buyers)
corr_item = pd.DataFrame(items_like_item, columns = ['Correlation'])
corr_item.dropna(inplace = True)
corr_item.head()
```

```python
# To get ten most similar items
corr_item.sort_values('Correlation', ascending=False).head()
```
