# Ex 4. For a given text file exclude the stop words and perform the Stemming & lemmatization and compare the results. 

```python
from nltk.tokenize import word_tokenize
from nltk.tag import pos_tag
from nltk.corpus import stopwords
from nltk import stem
from nltk.stem import PorterStemmer

ps = PorterStemmer()
wn_lemat = stem.WordNetLemmatizer()

stop_words = stopwords.words('english')

with open('data_in.txt',"r") as f:
    for sent in f:
        word_tokens=word_tokenize(sent)
        token=[]
        for word in word_tokens:
            token=token+[word]
       
        x = [word for word in token if word not in stop_words]
        
        words=[]
        words = [word for word in x if word.isalpha()]
        
        print(sent)
        print("{:10} {:10} {:10} ".format('Acutal Word |','Stemming |','Lemmatization'))
        
        for word in words:
            print("{:14} {:14} {:13} ".format(word, ps.stem(word), wn_lemat.lemmatize(word)))            
        print("--------------------------------------------\n")     
```

# Ex 5. Create a small dictionary file with required set of words with weightage attached to it with positive and negative numbers. Create a python script that analyzes the given text file and classify it as negative or positive sentiment.

```python
from nltk.tokenize import word_tokenize
```

```python
sent_dict = {"hate":-1,"hatred":-2,"annoyed":-2,"annoy":-1,"annoyingly":-2,"nasty":-2,"nice":1,"excellent":2,"good":1,"wonderful":2,"best":3,"better":2,"awesome":2,"beautiful":2,"beauty":1,"beautifully":2,"supreme":3}
```

```python
with open('NLPdataEx5data_senti_analyze.txt','r') as sent_analyze:
    for sent in sent_analyze:
        words=word_tokenize(sent)
        score=0
        for word in words:
            if(word.lower() in sent_dict.keys()):
                score=score+sent_dict[word]  

        sentiment=""
        if(score>0):
            sentiment='Positive'
        else:
            sentiment='Negative'
        print("{}\n ---->Score is {} and Sentiment is {}\n".format(sent,score,sentiment))
    
```
