```python
m1q= dataset['Price'].quantile(0.25)
m3q= dataset['Price'].quantile(0.75)

print (m1q,m3q)
```

```python
q90=dataset['Price'].quantile(0.90)
print (q90)

iqr=m3q-m1q
print ("Interquartile: ",iqr)

floor=(1.5*iqr)-m1q
print ("Floor: ",floor)

cap=(1.5*iqr)+m3q
print ("Cap: ",cap)
```

```python
#Outliner Treatment 

dataset_0=dataset[(dataset['Price']>=floor)&(dataset['Price']<=cap)]
```
